```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Función para cargar los datos desde un enlace URL
def cargar_datos(link):
    datos = pd.read_csv(link, index_col='id')
    return datos

# Función para aplicar LabelEncoder a las características categóricas
def aplicar_label_encoder(dataset):
    encoder = LabelEncoder()
    for columna in dataset.columns:
        if dataset[columna].dtype == 'object':  # Verificar si la columna es categórica
            dataset[columna] = encoder.fit_transform(dataset[columna])
    return dataset

# Función para aplicar la codificación de variables ficticias a las características categóricas
def aplicar_dummies(dataset):
    # Obtener las características categóricas
    categorical_features = ['proto', 'service', 'state']
    # Aplicar get_dummies a las características categóricas
    dataset = pd.get_dummies(dataset, columns=categorical_features)
    return dataset


```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Inicializar los LabelEncoders
le_service = LabelEncoder()
le_proto = LabelEncoder()
le_state = LabelEncoder()

# Codificar las características categóricas
datos_training['service_encoded'] = le_service.fit_transform(datos_training['service'])
datos_training['proto_encoded'] = le_proto.fit_transform(datos_training['proto'])
datos_training['state_encoded'] = le_state.fit_transform(datos_training['state'])

datos_testing['service_encoded'] = le_service.transform(datos_testing['service'])
datos_testing['proto_encoded'] = le_proto.transform(datos_testing['proto'])
datos_testing['state_encoded'] = le_state.transform(datos_testing['state'])

# Inicializar los OneHotEncoders
service_onehot = OneHotEncoder()
proto_onehot = OneHotEncoder()
state_onehot = OneHotEncoder()

# Codificar las características categóricas con One-Hot Encoding
X_train_service = service_onehot.fit_transform(datos_training['service_encoded'].values.reshape(-1, 1)).toarray()
X_train_proto = proto_onehot.fit_transform(datos_training['proto_encoded'].values.reshape(-1, 1)).toarray()
X_train_state = state_onehot.fit_transform(datos_training['state_encoded'].values.reshape(-1, 1)).toarray()

X_test_service = service_onehot.transform(datos_testing['service_encoded'].values.reshape(-1, 1)).toarray()
X_test_proto = proto_onehot.transform(datos_testing['proto_encoded'].values.reshape(-1, 1)).toarray()
X_test_state = state_onehot.transform(datos_testing['state_encoded'].values.reshape(-1, 1)).toarray()

# Concatenar las características codificadas one-hot con los conjuntos de datos originales
datos_training = pd.concat([datos_training, pd.DataFrame(X_train_service, columns=[f'service_{i}' for i in range(X_train_service.shape[1])]),
                            pd.DataFrame(X_train_proto, columns=[f'proto_{i}' for i in range(X_train_proto.shape[1])]),
                            pd.DataFrame(X_train_state, columns=[f'state_{i}' for i in range(X_train_state.shape[1])])], axis=1)

datos_testing = pd.concat([datos_testing, pd.DataFrame(X_test_service, columns=[f'service_{i}' for i in range(X_test_service.shape[1])]),
                           pd.DataFrame(X_test_proto, columns=[f'proto_{i}' for i in range(X_test_proto.shape[1])]),
                           pd.DataFrame(X_test_state, columns=[f'state_{i}' for i in range(X_test_state.shape[1])])], axis=1)

# Eliminar las columnas originales de características categóricas
datos_training.drop(columns=['service', 'proto', 'state', 'service_encoded', 'proto_encoded', 'state_encoded'], inplace=True)
datos_testing.drop(columns=['service', 'proto', 'state', 'service_encoded', 'proto_encoded', 'state_encoded'], inplace=True)

# Seleccionar características y clases de salida para el conjunto de datos de entrenamiento
X_train = datos_training.iloc[:, :-1]  # Excluir la última columna que contiene la etiqueta de clase
y_train = datos_training.iloc[:, -1]   # Seleccionar la última columna que contiene la etiqueta de clase

# Seleccionar características y clases de salida para el conjunto de datos de prueba
X_test = datos_testing.iloc[:, :-1]    # Excluir la última columna que contiene la etiqueta de clase
y_test = datos_testing.iloc[:, -1]     # Seleccionar la última columna que contiene la etiqueta de clase
```

```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model


# Seleccionar características basadas en la correlación de Spearman
corr_spearman = X_scaled_df.corr(method='spearman')
highly_correlated_features = set()

for i in range(len(corr_spearman.columns)):
    for j in range(i):
        if abs(corr_spearman.iloc[i, j]) > 0.8:  # Puedes ajustar este umbral según tus necesidades
            colname = corr_spearman.columns[i]
            highly_correlated_features.add(colname)

selected_features = [col for col in X_scaled_df.columns if col not in highly_correlated_features]

# Crear un nuevo DataFrame con las características seleccionadas
X_selected_features = X_scaled_df[selected_features]

# Imprimir las características seleccionadas
print(X_selected_features)

# Definir la arquitectura del autoencoder
input_dim = X_selected_features.shape[1]
encoding_dim = 10  # ajusta esta dimensión según tus necesidades

input_data = Input(shape=(input_dim,))
encoded = Dense(encoding_dim, activation='relu')(input_data)
decoded = Dense(input_dim, activation='sigmoid')(encoded)

autoencoder = Model(input_data, decoded)

# Compilar y entrenar el autoencoder
autoencoder.compile(optimizer='adam', loss='mean_squared_error')
autoencoder.fit(X_selected_features, X_selected_features, epochs=10, batch_size=32)

# Obtener las representaciones codificadas de los datos
encoded_data = autoencoder.predict(X_selected_features)

# Calcular la varianza de las representaciones codificadas
encoded_variance = encoded_data.var(axis=0)

# Imprimir la varianza de las representaciones codificadas
plt.figure(figsize=(8, 6))
plt.bar(range(len(encoded_variance)), encoded_variance, alpha=0.5, align='center', label='individual variance')
plt.legend()
plt.ylabel('Variance ratio')
plt.xlabel('Encoded components')
plt.show()
```
